```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Read the image:
img = cv2.imread("C:/Users/Phase/Downloads/Computer-Vision-with-Python/DATA/00-puppy.jpg")
plt.imshow(img)
```

```python
#Covert default color-space of computer vision module to actual color-space:
image = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
#Conversion the image into HSL and HSV:
hsv_image = cv2.cvtColor(image,cv2.COLOR_RGB2HSV)
plt.imshow(hsv_image)
```

```python
hls_image = cv2.cvtColor(image,cv2.COLOR_RGB2HLS)
plt.imshow(hls_image)
```
