```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import random
```

```python
def main():
    #Read the image:
    org_img = plt.imread("C:/Users/Phase/Downloads/Computer-Vision-with-Python/DATA/sammy.jpg")
    #org_img = cv2.cvtColor(org_img,cv2.COLOR_BGR2RGB)
    plt.imshow(org_img)
    
    #Calculate the shape of the image:
    rows,columns,channels = org_img.shape
    
    #Assigning image to another variable:
    out_img = org_img
    
    #Showing image in matplotlib window:
    plt.imshow(out_img)
    plt.title("Original Image")
    plt.show()

    #Let's define noise probabilty:
    noise_prob = 0.1  #10%

    #Create a placeholder for an output image:
    output_image = np.zeros(org_img.shape,np.uint8)

    #Let's define noise:
    #For this purpose we need 2 loops,one for the row and another for column:
    for i in range(rows):
        for j in range(columns):
        
            #Generate random number:
            rand_num = random.random()
        
            if rand_num < noise_prob/2:
            #Add pepper sprinkels:
                output_image[i][j] = [0,0,0]
            elif rand_num < noise_prob:
            #Add salt sprinkels:
                output_image[i][j] = [255,255,255]
            else:
                output_image[i][j] = org_img[i][j]
        
    #Display the output image:
    plt.imshow(output_image)
    plt.title("Image with Salt and Pepper noise")
    plt.show()
    
    #Remove the noise and display the image:
    #gaussian_img = cv2.GaussianBlur(output_image,(5,5),sigmaX=0)
    median_img = cv2.medianBlur(output_image,ksize=3)
    plt.title("Image after removing noise")
    plt.imshow(median_img)
```

```python
if __name__ == "__main__":
    main()
```
