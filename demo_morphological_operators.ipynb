```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Load the Image:
def load_img():
    img_canvas = np.zeros((600,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(img_canvas,text="ABCDE",org=(50,300),fontFace=font,fontScale=5,color=(255,255,255),thickness=25)
    return img_canvas
```

```python
#Display Image:
def disp_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
#Let's Start:
org_img = load_img()
disp_img(org_img)
```

```python
#Creating Kernel:
kernel = np.ones((5,5),dtype=np.uint8)
kernel
```

```python
#Calling "ERODE" function of cv2:EROSION ONLY
result_img = cv2.erode(org_img,kernel,iterations=3)
disp_img(result_img)
```

```python
#OPENING:EROSION followed by DILATION
#First let's add some white backgroud noise in the original image:
org_img = load_img()

#Add white backgroud noise:
white_noise = np.random.randint(low=0,high=2,size=(600,600))

#Display the noise:
#disp_img(white_noise)

#Scale the noise w.r.t. original image:
white_noise = white_noise * 255.0

#Let's make our original image noisy:
noisy_image = org_img + white_noise

#Let's see how noisy image look like:
disp_img(noisy_image)
```

```python
#Let's use opeing function to clear out this noise:
opening = cv2.morphologyEx(noisy_image,cv2.MORPH_OPEN,kernel)

#View the filtered image:
disp_img(opening)
```

```python
#Load the image:
org_img = load_img()

#Create Black Noise function:
black_noise = np.random.randint(low=0,high=2,size=(600,600))

#Scale the Noise function:
black_noise = black_noise * -255

#Add the noise into original image:
black_noise_image = org_img +black_noise

black_noise_image[black_noise_image == -255] = 0

#Display the image:
disp_img(black_noise_image)
```

```python
closing = cv2.morphologyEx(black_noise_image,cv2.MORPH_CLOSE,kernel)
disp_img(closing)
```
