#### INSERTION OF GAUSSIAN NOISE:

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
```

```python
#Read the Image:
input_image = cv2.imread("C:/ME-THESIS/IMAGE-DB/IMAGE-DB/standard_test_images/lena_color_512.tif")
input_image = cv2.cvtColor(input_image,cv2.COLOR_BGR2RGB)
plt.imshow(input_image)
plt.show()
```

```python
input_image
```

```python
#Add Gaussian Noise:
blurred_image = cv2.GaussianBlur(input_image,(5,5),2)

plt.subplot(121),plt.imshow(input_image),plt.title("Original Image")
plt.xticks([]),plt.yticks([])

plt.subplot(122),plt.imshow(blurred_image),plt.title("Filtered Image")
plt.xticks([]),plt.yticks([])

plt.show()
```

#### Effect of Standard Deviation(Sigma) on Gaussian Noise:

The magnitude of Gaussian Noise depends on the **Standard Deviation(sigma)**.Noise Magnitude is directly proportionl
to the **sigma value**:

```python
sigma_vals = [2,4,8,16,32]
for sigma in sigma_vals:
    blurred_image = cv2.GaussianBlur(input_image,(5,5),sigma)
    
    plt.subplot(121),plt.imshow(input_image),plt.title("Original Image")
    plt.xticks([]),plt.yticks([])

    plt.subplot(122),plt.imshow(blurred_image),plt.title("Filtered Image with Sigma:{}".format(sigma))
    plt.xticks([]),plt.yticks([])

    plt.show()
```
