```python
#Import required libraries:
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Read the images:
img1 = cv2.imread("C:/Users/Phase/Desktop/ME-THESIS-TEMPLATE/IMAGE-DB/standard_test_images/lena_color_512.tif")
img2 = cv2.imread("C:/Users/Phase/Desktop/ME-THESIS-TEMPLATE/IMAGE-DB/standard_test_images/mandril_color.tif")
```

```python
#Covert the image into RGB form:
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)

#Show the converted image:
plt.imshow(img2)
```

```python
#Image Addition:
img_add = img1 + img2

#Image Substration:
sub_1 = img1 - img2
sub_2 = img2 - img1
```

```python
image_titles = ['Lena','Mandril','img_add','img1 - img2','img2 - img1']
images = [img1,img2,img_add,sub_1,sub_2]

#Plot the images:
for i in range(5):
    plt.subplot(2,5,i+1)
    plt.imshow(images[i])
    plt.title(image_titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
